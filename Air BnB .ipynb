```python
#----finding outliers-----
```

```python
def outliers(df,price):
    qrt1=df['price'].quantile(0.25)
    qrt3=df['price'].quantile(0.75)
    iqr=qrt3-qrt1
    l_bound=qrt1-1.5*iqr
    u_bound=qrt3+1.5*iqr
    return df[(df['price']<l_bound) |(df['price']>u_bound)]


outliers_count=len(outliers(df,'price'))
print("outliers in Prices are: ",outliers_count)

```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['price'], color='skyblue')
plt.title("Airbnb Outlier Prices")
plt.xlabel("Price ($)")
plt.show()
```

```python
#-----dropping outlier values----
```

```python
qrt1=df['price'].quantile(0.25)
qrt3=df['price'].quantile(0.75)
iqr=qrt3-qrt1
l_bound=qrt1-1.5*iqr
u_bound=qrt3+1.5*iqr
outlier_indices = df[(df['price']<l_bound) | (df['price']>u_bound)].index
df_cleaned=df.drop(outlier_indices)
df_cleaned.drop((['last_review','reviews_per_month']),axis=1,inplace=True)
```

```python
df_cleaned.info()
```

```python
#----Exploratory Data Analysis (EDA)-----
```

```python
#--Price Distribution Analsys----
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df_cleaned['price'],bins=50,kde=True,color='coral')
plt.title("Distribution of Listing Price")
plt.xlabel("Price in $")
plt.show()
```

```python
#----Neighborhood Comparison------
```

```python
neigh_stats=df_cleaned.groupby('neighbourhood')[['price','number_of_reviews','minimum_nights']].agg({'price':'mean','number_of_reviews':'sum','minimum_nights':'count'}).sort_values('price',ascending=False)
print(neigh_stats)
plt.figure(figsize=(15,12))
neigh_stats.head(5).plot(kind='bar',secondary_y='number_of_reviews')
plt.title('Price vs. Popularity by Neighborhood')
plt.xlabel("Neighbourhood")
plt.show()
```
